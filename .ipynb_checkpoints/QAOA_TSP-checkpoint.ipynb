```python
import projectq as pq
from projectq import MainEngine
from projectq.ops import All, Measure, QubitOperator, TimeEvolution, X, H

import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import minimize_scalar

from classical_TSP.random_graph_generator import gen_graph,graphs_decipher
from classical_TSP.tsp_dp_solver import get_distance_matrix
from graph_converter import graph_to_hamiltonian

from QAOA import QAOA
```

```python
target_file="./classical_TSP/random_graphs.csv"
np.loadtxt(target_file)[:10]
```

```python
# analysis terms in cost and panilty Hamiltonian 
n_terms=[]
max_n=10
for n in range(1,max_n):
    target_file="./classical_TSP/random_graphs.csv"
    g=graphs_decipher(target_file,n)
    distance_matrix=get_distance_matrix(g[0])
    cost_H=graph_to_hamiltonian(distance_matrix)
    n_terms.append(len(cost_H.terms))
    print("Vertice: {0} Hamiltonian Terms:{1}".format(n,n_terms[-1]))
```

```python
f=plt.figure()
f.dpi=80
plt.xlabel("Size of graph nodes")
plt.ylabel(r"Single-qubit gates in $\hat{H}_{cost}$")
plt.plot(n_terms,'o--')
```

```python
f=plt.figure()
f.dpi=80
plt.xlabel("Size of graph nodes")
plt.ylabel(r"Single-qubit gates in $\hat{H}_{cost}$ [Log Scale]")
plt.plot(np.log(n_terms),'o--')
```

```python
eng = MainEngine()
maxcut_solver = QAOA(cost = cost_H,
                     mixer = None,
                     n_steps = 1, 
                     engine = eng, 
                     vqe_run_kwargs = {'verbose': 1, 'draws': 10})
```

```python
betas, gammas = maxcut_solver.solve_angles()
print("Values of betas:", betas)
print("Values of gammas:", gammas)
```
